```python
import os
import json
from pprint import pprint
```

```python
filepath = os.path.join("factbook.json")
with open(filepath, encoding="utf8") as jsonfile:
    cia_json = json.load(jsonfile)
```

```python
poverty = {}
genderratio = {}

for country in cia_json['countries'].keys():
    
    try:
        poverty[cia_json['countries'][country]['data']['name']] = cia_json['countries'][country]['data']['economy']['population_below_poverty_line']['value']
        genderratio[cia_json['countries'][country]['data']['name']] = round(cia_json['countries'][country]['data']['people']['sex_ratio']['total_population']['value']/(cia_json['countries'][country]['data']['people']['sex_ratio']['total_population']['value']+1),4)
        
    except Exception as ex:
        poverty[cia_json['countries'][country]['data']['name']] = ''
        genderratio[cia_json['countries'][country]['data']['name']] = ''
```
